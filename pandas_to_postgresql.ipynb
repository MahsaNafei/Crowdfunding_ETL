```python
# Import dependecies
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
```

```python
# Read contacts csv
contacts_df = pd.read_csv("../Crowdfunding_ETL/Output/contacts.csv")
contacts_df
```

```python
# Create engine for SQL
engine = create_engine('postgresql://postgres:postgres@localhost:5432/df')
```

```python
# import to sql
contacts_df.to_sql("contacts", con=engine.connect(), if_exists='replace')
```

```python
contacts_df = pd.read_sql('SELECT * FROM contacts', con=engine.connect())
contacts_df.head(10)
```
